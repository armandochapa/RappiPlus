# Proyecto RappiPlus: de datos a decisiones de negocio

**Introducción**


El objetivo de este proyecto es evaluar el desempeño del servicio **RappiPlus** para apoyar **decisiones de negocio basadas en datos**.

Se trabajan con múltiples datasets del negocio:

- **rappiplus_orders_raw.csv** → información de pedidos, precios, descuentos y revenue  
- **rappiplus_catalog.csv** → costos de productos, categorías y proveedores  
- **rappiplus_marketing_spend.csv** → inversión en marketing por canal y país  
- **events / users / user_activity (SQL)** → comportamiento del usuario dentro de la plataforma  
- **experiment_checkout_ui.csv** → resultados de un experimento A/B en el checkout  

El análisis sigue una lógica clara y progresiva:

1. 🔍 Evaluar si podemos confiar en los datos (calidad de datos en Python) 

2. 💰 Analizar si el negocio es rentable (revenue, costos y profit)  

3. 🛒 Entender dónde se pierden los usuarios (funnel de conversión)  

4. 🔁 Evaluar si los usuarios regresan (retención por cohortes)  

5. 🧪 Validar si los cambios generan impacto (test estadístico)  

6. 📊 Comunicar los resultados (dashboard en BI)  

A lo largo del proyecto, se transforman datos en insights para responder preguntas clave del negocio y proponer **recomendaciones accionables**.

### Revisión y calidad de datos

**🎯 Objetivo:** Detectar y corregir problemas en los datos que puedan afectar el análisis de revenue, costos y rentabilidad.

Se revisan los 3 datasets

```python
# tu código aquí

print("@@@@@@@@@@ Limpieza de tabla orders @@@@@@@@@@")
print("")

print("######################################################")
print("# Convertir fechas al formato correcto")
print("######################################################")
print("")
orders["fecha_hora_pedido"] = pd.to_datetime(orders["fecha_hora_pedido"], errors="coerce")
orders.info()
print("")
orders["cantidad"].describe()
print("")

print("######################################################")
print("# Revisar variables numéricas (sin negativos o ceros inválidos)")
print("######################################################")
print("")

# Filtrar solo registros válidos
orders_clean = orders[
    (orders['cantidad'] > 0) & 
    (orders['monto_total'] > 0)
].copy()

# Analizar qué registros se están eliminando
registros_problematicos = orders[
    ~((orders['cantidad'] > 0) & (orders['monto_total'] > 0))
]

# Arreglo de variables numéricas (sin negativos o ceros inválidos) y consistencia de montos
print(f"Total de registros problemáticos: {len(registros_problematicos)}")
print("\nDesglose de problemas:")
print("Cantidad <= 0:", (orders['cantidad'] <= 0).sum())
print("Cantidad NaN:", orders['cantidad'].isna().sum())
print("Monto <= 0:", (orders['monto_total'] <= 0).sum())
print("Monto NaN:", orders['monto_total'].isna().sum())
print("")

print(f"Registros eliminados: {len(orders) - len(orders_clean)}")
print("")

print("######################################################")
print("# Verificar consistencia de montos")
print("######################################################")
print("")

# Verificar si los montos son consistentes
orders_clean['monto_calculado'] = (orders_clean['cantidad'] * orders_clean['precio_unitario']) - orders_clean['monto_descuento']

# Comparar con el monto_total registrado
orders_clean['diferencia'] = orders_clean['monto_total'] - orders_clean['monto_calculado']

# Encontrar inconsistencias
inconsistencias = orders_clean[orders_clean['diferencia'].abs() > 0.02]  # Tolerancia de 2 centavos

print(f"Registros inconsistentes: {len(inconsistencias)}")
print("")

print("######################################################")
print("# Eliminar duplicados")
print("######################################################")
print("")

# Detectar duplicados exactos (toda la fila idéntica)
duplicados_exactos = orders_clean.duplicated().sum()
print(f"Duplicados exactos: {duplicados_exactos}")

# Detectar duplicados por ID de pedido (más importante)
duplicados_pedido = orders_clean.duplicated(subset=['id_pedido']).sum()
print(f"Duplicados por ID de pedido: {duplicados_pedido}")

# Ver ejemplos de duplicados por ID de pedido
if duplicados_pedido > 0:
    print("\nEjemplos de pedidos duplicados:")
    pedidos_duplicados = orders_clean[orders_clean.duplicated(subset=['id_pedido'], keep=False)]
    print(pedidos_duplicados[['id_pedido', 'id_usuario', 'fecha_hora_pedido', 'monto_total']].head(10))
print("")


# Analizar los duplicados más detalladamente
print("Análisis detallado de duplicados:")
pedidos_duplicados = orders_clean[orders_clean.duplicated(subset=['id_pedido'], keep=False)]
pedidos_duplicados_sorted = pedidos_duplicados.sort_values('id_pedido')

# Ver un ejemplo específico
ejemplo_pedido = pedidos_duplicados_sorted[pedidos_duplicados_sorted['id_pedido'] == 'order_734']
print(f"\nEjemplo del pedido order_734:")
print(ejemplo_pedido[['id_pedido', 'id_usuario', 'fecha_hora_pedido', 'monto_total', 'cantidad', 'precio_unitario']])

# Contar cuántos duplicados tiene cada pedido
duplicados_count = pedidos_duplicados.groupby('id_pedido').size()
print(f"\nDistribución de duplicados:")
print(f"Pedidos con 2 registros: {(duplicados_count == 2).sum()}")
print(f"Pedidos con 3+ registros: {(duplicados_count > 2).sum()}")
print("")

# Antes de eliminar, guardemos información sobre los duplicados
print("=== ANTES DE ELIMINAR DUPLICADOS ===")
print(f"Total de registros: {len(orders_clean)}")
print(f"Pedidos únicos: {orders_clean['id_pedido'].nunique()}")
print(f"Duplicados por ID de pedido: {orders_clean.duplicated(subset=['id_pedido']).sum()}")

# Eliminar duplicados manteniendo el primer registro de cada id_pedido
orders_clean_no_duplicates = orders_clean.drop_duplicates(subset=['id_pedido'], keep='first')

# Verificar los resultados
print("\n=== DESPUÉS DE ELIMINAR DUPLICADOS ===")
print(f"Total de registros: {len(orders_clean_no_duplicates)}")
print(f"Pedidos únicos: {orders_clean_no_duplicates['id_pedido'].nunique()}")
print(f"Duplicados restantes: {orders_clean_no_duplicates.duplicated(subset=['id_pedido']).sum()}")

# Calcular cuántos registros eliminamos
registros_eliminados = len(orders_clean) - len(orders_clean_no_duplicates)
print(f"\nRegistros eliminados: {registros_eliminados}")
print("")

print("######################################################")
print("# Revisar variables categóricas")
print("######################################################")
print("")

# Identificar variables categóricas en el dataset
print("=== IDENTIFICACIÓN DE VARIABLES CATEGÓRICAS ===")
categorical_columns = orders_clean_no_duplicates.select_dtypes(include=['object']).columns.tolist()
print(f"Variables categóricas encontradas: {categorical_columns}")
print(f"Total de variables categóricas: {len(categorical_columns)}")

print("\n" + "="*60)
print("")

# Revisar cada variable categórica
for column in categorical_columns:
    print(f"\n=== ANÁLISIS DE: {column.upper()} ===")
    
    # Información básica
    #print(f"Valores únicos: {orders_clean_no_duplicates[column].nunique()}")
    print(f"Valores nulos: {orders_clean_no_duplicates[column].isnull().sum()}")
    
    # Mostrar todos los valores únicos
    #print(f"\nValores únicos en '{column}':")
    unique_values = orders_clean_no_duplicates[column].unique()
    #for value in sorted(unique_values):
    #    print(f"  - {value}")
    
    # Distribución de frecuencias
    #print(f"\nDistribución de frecuencias:")
    #value_counts = orders_clean_no_duplicates[column].value_counts()
    #for value, count in value_counts.items():
    #    percentage = (count / len(orders_clean_no_duplicates)) * 100
    #    print(f"  {value}: {count:,} ({percentage:.1f}%)")
    
    print("-" * 40)
print("")

# Investigar patrones de valores nulos
print("=== ANÁLISIS DE PATRONES DE VALORES NULOS ===")

# Verificar si los nulos coinciden en las mismas filas
null_pattern = orders_clean_no_duplicates[['pais', 'dispositivo', 'fuente_referencia', 
                                          'nombre_producto', 'categoria_producto']].isnull()

print("Registros con múltiples valores nulos:")
multiple_nulls = null_pattern.sum(axis=1) > 1
print(f"Filas con más de 1 valor nulo: {multiple_nulls.sum()}")

# Ver qué combinaciones de nulos tenemos
print("\nCombinaciones de valores nulos:")
null_combinations = null_pattern.groupby(list(null_pattern.columns)).size()
print(null_combinations[null_combinations > 0])

# Verificar si los 30 nulos son los mismos registros
print("\n=== VERIFICACIÓN DE REGISTROS CON NULOS ===")
registros_con_nulos_producto = orders_clean_no_duplicates['nombre_producto'].isnull()
registros_con_nulos_categoria = orders_clean_no_duplicates['categoria_producto'].isnull()
registros_con_nulos_fuente = orders_clean_no_duplicates['fuente_referencia'].isnull()

print(f"¿Los nulos de producto y categoría son los mismos? {registros_con_nulos_producto.equals(registros_con_nulos_categoria)}")
print(f"¿Los nulos de fuente y producto son los mismos? {registros_con_nulos_fuente.equals(registros_con_nulos_producto)}")
print("")

# Analizar el impacto de los registros con nulos múltiples
print("=== ANÁLISIS DE IMPACTO DE REGISTROS CON NULOS ===")

# Registros con nulos múltiples (los 30 problemáticos)
registros_nulos_multiples = orders_clean_no_duplicates[
    orders_clean_no_duplicates['fuente_referencia'].isnull() &
    orders_clean_no_duplicates['nombre_producto'].isnull() &
    orders_clean_no_duplicates['categoria_producto'].isnull()
]

print(f"Registros con nulos múltiples: {len(registros_nulos_multiples)}")
print(f"Revenue de estos registros: ${registros_nulos_multiples['monto_total'].sum():,.2f}")
print(f"% del revenue total: {(registros_nulos_multiples['monto_total'].sum() / orders_clean_no_duplicates['monto_total'].sum()) * 100:.2f}%")

# Mostrar algunos ejemplos
print("\nEjemplos de registros problemáticos:")
print(registros_nulos_multiples[['id_pedido', 'pais', 'monto_total', 'fuente_referencia', 'nombre_producto']].head())


# Detectar posibles problemas en variables categóricas
print("\n=== DETECCIÓN DE PROBLEMAS POTENCIALES ===")

for column in categorical_columns:
    print(f"\n--- Revisando {column} ---")
    
    # Detectar valores con espacios extra
    values_with_spaces = orders_clean_no_duplicates[column].str.contains(r'^\s+|\s+$', na=False).sum()
    if values_with_spaces > 0:
        print(f"⚠️  Valores con espacios extra: {values_with_spaces}")
    
    # Detectar valores muy similares (posibles errores de escritura)
    unique_values = orders_clean_no_duplicates[column].dropna().unique()
    if len(unique_values) > 1:
        for i, val1 in enumerate(unique_values):
            for val2 in unique_values[i+1:]:
                if isinstance(val1, str) and isinstance(val2, str):
                    if val1.lower() == val2.lower() and val1 != val2:
                        print(f"⚠️  Posible inconsistencia: '{val1}' vs '{val2}'")
    
    # Detectar valores poco frecuentes (posibles errores)
    value_counts = orders_clean_no_duplicates[column].value_counts()
    rare_values = value_counts[value_counts == 1]
    if len(rare_values) > 0:
        print(f"⚠️  Valores que aparecen solo 1 vez: {len(rare_values)}")
        if len(rare_values) <= 5:  # Mostrar solo si son pocos
            print(f"     Valores: {list(rare_values.index)}")


#######################################################

# Crear una copia explícita del DataFrame
print("=== CORRECCIÓN DE INCONSISTENCIAS EN PAÍS (MÉTODO SEGURO) ===")

# Crear una copia explícita para evitar el warning
orders_final = orders_clean_no_duplicates.copy()

# Ver los valores únicos antes de la corrección
print("Valores únicos en 'pais' ANTES de la corrección:")
print(orders_final['pais'].value_counts())

# Estandarizar los nombres de países (primera letra mayúscula)
orders_final['pais'] = orders_final['pais'].str.title()

# Verificar la corrección
print("\nValores únicos en 'pais' DESPUÉS de la corrección:")
print(orders_final['pais'].value_counts())

# Verificar que no hay más inconsistencias
print(f"\nPaíses únicos después de la corrección: {orders_final['pais'].nunique()}")

#######################################################
#######################################################
#######################################################

# Resumen:

# Análisis de limpieza actual

# Lo que ya está implementado:
#    - Conversión de fechas a formato datetime
#    - Eliminación de registros con valores negativos o nulos en variables críticas
#    - Verificación de consistencia de montos
#    - Eliminación de duplicados por ID de pedido
#    - Estandarización de nombres de países
#    - Análisis detallado de patrones de valores nulos

# MANTENER (por ahora):
#1. Los 30 registros con múltiples nulos
#    - Solo representan 0.02% del revenue ($12,262.62)
#    - Tienen información válida de país y monto_total
#    - Útiles para análisis de revenue por país

#2. Los 296 registros con país nulo
#    - Tienen información completa de productos y revenue
#    - Útiles para análisis de productos y categorías

#3. Los 20 registros con dispositivo nulo
#    - Información menor, no afecta análisis principales

# ¿Por qué esta estrategia?
#     Conservar el 99.8% del revenue para análisis de rentabilidad
#     Mantener flexibilidad para análisis específicos
#     Poder filtrar según la necesidad en cada análisis

```

```python
# Explorar tabla events
# =========================
query_events = '''
SELECT *
FROM events;
'''
events = pd.read_sql(query_events, con=engine)
events.head()
```

```python
# PARTE 1: Totales del funnel
# ======================

query_totals = '''
SELECT 
    nombre_evento,
    COUNT(DISTINCT id_usuario) as usuarios_unicos
FROM events
GROUP BY nombre_evento
ORDER BY 
    CASE nombre_evento
        WHEN 'first_visit' THEN 1
        WHEN 'add_to_cart' THEN 2
        WHEN 'select_item' THEN 3
        WHEN 'begin_checkout' THEN 4
        WHEN 'add_payment_info' THEN 5
        WHEN 'purchase' THEN 6
        ELSE 7
    END;
'''

totals = pd.read_sql(query_totals, con=engine)
totals
```

```python
# PARTE 2: Conversiones
# ======================

query_conversion = '''
WITH funnel_ordenado AS (
    SELECT 
        nombre_evento,
        COUNT(DISTINCT id_usuario) as usuarios_unicos,
        CASE nombre_evento
            WHEN 'first_visit' THEN 1
            WHEN 'add_to_cart' THEN 2
            WHEN 'select_item' THEN 3
            WHEN 'begin_checkout' THEN 4
            WHEN 'add_payment_info' THEN 5
            WHEN 'purchase' THEN 6
            ELSE 7
        END as orden_etapa
    FROM events
    GROUP BY nombre_evento
),
funnel_con_conversion AS (
    SELECT 
        nombre_evento,
        usuarios_unicos,
        orden_etapa,
        LAG(usuarios_unicos) OVER (ORDER BY orden_etapa) as usuarios_etapa_anterior
    FROM funnel_ordenado
)
SELECT 
    nombre_evento,
    usuarios_unicos,
    usuarios_etapa_anterior,
    CASE 
        WHEN usuarios_etapa_anterior IS NOT NULL 
        THEN ROUND((usuarios_unicos::DECIMAL / usuarios_etapa_anterior) * 100, 2)
        ELSE 100.0 
    END as tasa_conversion_pct,
    CASE 
        WHEN usuarios_etapa_anterior IS NOT NULL 
        THEN usuarios_etapa_anterior - usuarios_unicos
        ELSE 0 
    END as usuarios_perdidos
FROM funnel_con_conversion
ORDER BY orden_etapa;
'''

conversion = pd.read_sql(query_conversion, con=engine)
conversion
```

```python
# PARTE 2: Tasa de Conversión final
# ======================

query_conversion_final = '''
WITH conversion_final AS (
    SELECT 
        SUM(CASE WHEN nombre_evento = 'first_visit' THEN 1 ELSE 0 END) as visitantes_iniciales,
        SUM(CASE WHEN nombre_evento = 'purchase' THEN 1 ELSE 0 END) as compradores_finales
    FROM (
        SELECT DISTINCT id_usuario, nombre_evento 
        FROM events 
        WHERE nombre_evento IN ('first_visit', 'purchase')
    ) t
)
SELECT 
    visitantes_iniciales,
    compradores_finales,
    ROUND((compradores_finales::DECIMAL / visitantes_iniciales) * 100, 2) as conversion_final_pct
FROM conversion_final;
'''

conversion_final = pd.read_sql(query_conversion_final, con=engine)
conversion_final
```

## 🔹 Paso 4: Evaluar si los usuarios regresan (retención por cohortes)

**🎯 Objetivo:** Analizar la retención de usuarios para entender si regresan después de registrarse.

**Tablas**

- `users` 
- `user_activity` 

---
1. Se identifica la cohorte de cada usuario según el **mes de registro**.


2. Se calcula la retención semanal: cuántos usuarios **se mantienen activos** en cada semana desde su registro.
   - `retenido_w1`: usuarios activos en la semana 1  
   - `retenido_w2`: usuarios activos en la semana 2  
   - `retenido_w3`: usuarios activos en la semana 3  


3. Se calcula el porcentaje de retención para cada semana, dividiendo los usuarios retenidos entre los clientes iniciales de la cohorte:  
   - `semana_1`: porcentaje de usuarios retenidos en la semana 1  
   - `semana_2`: porcentaje de usuarios retenidos en la semana 2  
   - `semana_3`: porcentaje de usuarios retenidos en la semana 3  

Se revisa que la columna de fecha esté en formato correcto (`DATE`).  
Se realiza la conversión usando: `CAST(fecha_registro AS DATE)`

```python
# Explorar tabla users
# =========================
query_users = '''
SELECT *
FROM users;
'''
users = pd.read_sql(query_users, con=engine)
users.head(3)
```

```python
# Explorar tabla user_activity
# =========================
query_user_activity = '''
SELECT *
FROM user_activity;
'''
user_activity = pd.read_sql(query_user_activity, con=engine)
user_activity.head(3)
```

```python
# 1: Identificar cohortes por mes de registro
# =========================
query_cohortes_mes = '''
WITH cohortes_usuarios AS (
    SELECT 
        id_usuario,
        CAST(fecha_registro AS DATE) as fecha_registro,
        DATE_TRUNC('month', CAST(fecha_registro AS DATE)) as cohorte_mes
    FROM users
)
SELECT 
    cohorte_mes,
    COUNT(DISTINCT id_usuario) as usuarios_registrados
FROM cohortes_usuarios
GROUP BY cohorte_mes
ORDER BY cohorte_mes;
'''
cohortes_mes = pd.read_sql(query_cohortes_mes, con=engine)
cohortes_mes
```

```python
# 2: Calcular retención semanal por cohorte
# =========================
query_retencion_semanal = '''
WITH cohortes_usuarios AS (
    SELECT 
        id_usuario,
        CAST(fecha_registro AS DATE) as fecha_registro,
        DATE_TRUNC('month', CAST(fecha_registro AS DATE)) as cohorte_mes
    FROM users
),
actividad_semanal AS (
    SELECT 
        cu.cohorte_mes,
        cu.id_usuario,
        ua.dias_despues_registro,
        ua.activo,
        CASE 
            WHEN ua.dias_despues_registro BETWEEN 1 AND 7 THEN 'semana_1'
            WHEN ua.dias_despues_registro BETWEEN 8 AND 14 THEN 'semana_2'
            WHEN ua.dias_despues_registro BETWEEN 15 AND 21 THEN 'semana_3'
            ELSE 'otras_semanas'
        END as periodo_semana
    FROM cohortes_usuarios cu
    JOIN user_activity ua ON cu.id_usuario = ua.id_usuario
    WHERE ua.dias_despues_registro <= 21
),
usuarios_por_cohorte AS (
    SELECT 
        cohorte_mes,
        COUNT(DISTINCT id_usuario) as usuarios_iniciales
    FROM cohortes_usuarios
    GROUP BY cohorte_mes
),
retencion_por_semana AS (
    SELECT 
        cohorte_mes,
        periodo_semana,
        COUNT(DISTINCT CASE WHEN activo = 1 THEN id_usuario END) as usuarios_retenidos
    FROM actividad_semanal
    WHERE periodo_semana IN ('semana_1', 'semana_2', 'semana_3')
    GROUP BY cohorte_mes, periodo_semana
)
SELECT 
    upc.cohorte_mes,
    upc.usuarios_iniciales,
    COALESCE(MAX(CASE WHEN rps.periodo_semana = 'semana_1' THEN rps.usuarios_retenidos END), 0) as retenido_w1,
    COALESCE(MAX(CASE WHEN rps.periodo_semana = 'semana_2' THEN rps.usuarios_retenidos END), 0) as retenido_w2,
    COALESCE(MAX(CASE WHEN rps.periodo_semana = 'semana_3' THEN rps.usuarios_retenidos END), 0) as retenido_w3,
    ROUND(
        COALESCE(MAX(CASE WHEN rps.periodo_semana = 'semana_1' THEN rps.usuarios_retenidos END), 0) * 100.0 / upc.usuarios_iniciales, 2
    ) as semana_1_pct,
    ROUND(
        COALESCE(MAX(CASE WHEN rps.periodo_semana = 'semana_2' THEN rps.usuarios_retenidos END), 0) * 100.0 / upc.usuarios_iniciales, 2
    ) as semana_2_pct,
    ROUND(
        COALESCE(MAX(CASE WHEN rps.periodo_semana = 'semana_3' THEN rps.usuarios_retenidos END), 0) * 100.0 / upc.usuarios_iniciales, 2
    ) as semana_3_pct
FROM usuarios_por_cohorte upc
LEFT JOIN retencion_por_semana rps ON upc.cohorte_mes = rps.cohorte_mes
GROUP BY upc.cohorte_mes, upc.usuarios_iniciales
ORDER BY upc.cohorte_mes;
'''
retencion_semanal = pd.read_sql(query_retencion_semanal, con=engine)
retencion_semanal
```

```python
# 3: Retención por cohortes
# ======================

query_cohort_retention_final = '''
WITH cohortes_usuarios AS (
    SELECT 
        id_usuario,
        CAST(fecha_registro AS DATE) as fecha_registro,
        DATE_TRUNC('month', CAST(fecha_registro AS DATE)) as cohorte_mes
    FROM users
),
usuarios_por_cohorte AS (
    SELECT 
        cohorte_mes,
        COUNT(DISTINCT id_usuario) as usuarios_iniciales
    FROM cohortes_usuarios
    GROUP BY cohorte_mes
),
actividad_semanal AS (
    SELECT 
        cu.cohorte_mes,
        cu.id_usuario,
        ua.dias_despues_registro,
        ua.activo,
        CASE 
            WHEN ua.dias_despues_registro BETWEEN 1 AND 7 THEN 'semana_1'
            WHEN ua.dias_despues_registro BETWEEN 8 AND 14 THEN 'semana_2'
            WHEN ua.dias_despues_registro BETWEEN 15 AND 21 THEN 'semana_3'
        END as periodo_semana
    FROM cohortes_usuarios cu
    JOIN user_activity ua ON cu.id_usuario = ua.id_usuario
    WHERE ua.dias_despues_registro <= 21 AND ua.activo = 1
),
retencion_por_semana AS (
    SELECT 
        cohorte_mes,
        periodo_semana,
        COUNT(DISTINCT id_usuario) as usuarios_retenidos
    FROM actividad_semanal
    WHERE periodo_semana IN ('semana_1', 'semana_2', 'semana_3')
    GROUP BY cohorte_mes, periodo_semana
)
SELECT 
    upc.cohorte_mes,
    upc.usuarios_iniciales,
    ROUND(
        COALESCE(MAX(CASE WHEN rps.periodo_semana = 'semana_1' THEN rps.usuarios_retenidos END), 0) * 100.0 / upc.usuarios_iniciales, 2
    ) as semana_1,
    ROUND(
        COALESCE(MAX(CASE WHEN rps.periodo_semana = 'semana_2' THEN rps.usuarios_retenidos END), 0) * 100.0 / upc.usuarios_iniciales, 2
    ) as semana_2,
    ROUND(
        COALESCE(MAX(CASE WHEN rps.periodo_semana = 'semana_3' THEN rps.usuarios_retenidos END), 0) * 100.0 / upc.usuarios_iniciales, 2
    ) as semana_3
FROM usuarios_por_cohorte upc
LEFT JOIN retencion_por_semana rps ON upc.cohorte_mes = rps.cohorte_mes
GROUP BY upc.cohorte_mes, upc.usuarios_iniciales
ORDER BY upc.cohorte_mes;
'''

# Ejecutar la consulta
cohorte_final = pd.read_sql(query_cohort_retention_final, con=engine)
cohorte_final
```

## 🔹 Paso 5: Validar si los cambios generan impacto (test estadístico)

🎯 **Objetivo:** Evaluar si la modificación en la UI del checkout impacta la **tasa de conversión de compra**.

---

1. **Analizar el dataset** `experiment_checkout_ui.csv` para identificar la métrica principal **conversion**.
   - La métrica **conversion** es 1 si el usuario completó la compra, 0 si no.    
2. **Plantear la hipótesis estadística**     
3. **Aplicar el test estadístico adecuado** 
4. **Interpretar el resultado**  

---
Hipótesis estadística
   - **H₀ (Hipótesis nula):** La tasa de conversión es igual en ambas páginas. Cualquier diferencia se debe al azar del muestreo.
   - **H₁ (Hipótesis alternativa):** La tasa de conversión es diferente entre ambas páginas. Hay un factor que influye en la decisión de compra.
   
**Test estadístico:** ...  
**Nivel de significancia alpha:** ...

```python
# tu código aquí
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind, levene, chi2_contingency
from statsmodels.stats.proportion import proportions_ztest
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar datos
df = pd.read_csv('https://practicum-content.s3.amazonaws.com/datasets/experiment_checkout_ui.csv')

# Definir nivel de significancia
alpha = 0.05
print(f"Nivel de significancia (α): {alpha}")
print(f"Nivel de confianza: {(1-alpha)*100}%")
```

```python
# Análisis de Tasa de Conversión (Prueba Z)
# Contar conversiones y totales por variante
conversiones = df.groupby('variante')['convirtio'].sum()
totales = df.groupby('variante')['convirtio'].count()

print("=== ANÁLISIS DE CONVERSIÓN POR VARIANTE ===")
print(f"Control - Conversiones: {conversiones['control']} de {totales['control']}")
print(f"Tratamiento - Conversiones: {conversiones['tratamiento']} de {totales['tratamiento']}")

# Calcular tasas de conversión
tasa_control = conversiones['control'] / totales['control']
tasa_tratamiento = conversiones['tratamiento'] / totales['tratamiento']

print(f"\nTasa de conversión Control: {tasa_control:.4f} ({tasa_control:.2%})")
print(f"Tasa de conversión Tratamiento: {tasa_tratamiento:.4f} ({tasa_tratamiento:.2%})")
print(f"Diferencia absoluta: {abs(tasa_tratamiento - tasa_control):.4f} ({abs(tasa_tratamiento - tasa_control):.2%})")

# Preparar datos para la prueba z
exitos = [conversiones['control'], conversiones['tratamiento']]
observaciones = [totales['control'], totales['tratamiento']]

# Aplicar prueba z de proporciones
z_stat, p_value_conversion = proportions_ztest(exitos, observaciones)

print(f"\n=== PRUEBA Z DE PROPORCIONES ===")
print(f"Estadístico z: {z_stat:.4f}")
print(f"Valor p: {p_value_conversion:.6f}")

# Interpretar resultados
if p_value_conversion < alpha:
    print(f"- Rechazamos H₀: Hay evidencia de diferencia significativa (p < {alpha})")
    mejor_variante = "Tratamiento" if tasa_tratamiento > tasa_control else "Control"
    print(f"   La variante {mejor_variante} tiene mejor conversión")
else:
    print(f"- No rechazamos H₀: No hay evidencia suficiente de diferencia (p >= {alpha})")
    print("   Las tasas de conversión son estadísticamente similares")
```

```python
# Análisis Adicional: Prueba Chi-cuadrado
# Crear tabla de contingencia
tabla_contingencia = pd.crosstab(df['variante'], df['convirtio'], margins=True)
print(f"\n=== TABLA DE CONTINGENCIA ===")
print(tabla_contingencia)

# Aplicar prueba Chi-cuadrado (sin los totales)
tabla_sin_totales = pd.crosstab(df['variante'], df['convirtio'])
chi2_stat, p_value_chi2, dof, expected = chi2_contingency(tabla_sin_totales)

print(f"\n=== PRUEBA CHI-CUADRADO ===")
print(f"Estadístico Chi-cuadrado: {chi2_stat:.4f}")
print(f"Valor p: {p_value_chi2:.6f}")
print(f"Grados de libertad: {dof}")

# Mostrar frecuencias esperadas
print(f"\n=== FRECUENCIAS ESPERADAS ===")
expected_df = pd.DataFrame(expected, 
                          index=tabla_sin_totales.index, 
                          columns=tabla_sin_totales.columns)
print(expected_df.round(2))

# Interpretar resultados
print(f"\n=== INTERPRETACIÓN CHI-CUADRADO ===")
if p_value_chi2 < alpha:
    print(f"- Rechazamos H₀: Existe asociación significativa entre variante y conversión (p < {alpha})")
    print("  Las variables NO son independientes")
else:
    print(f"- No rechazamos H₀: No hay evidencia de asociación significativa (p >= {alpha})")
    print("  Las variables son independientes")

# Verificar supuestos de la prueba Chi-cuadrado
print(f"\n=== VERIFICACIÓN DE SUPUESTOS ===")
print("Frecuencias esperadas mínimas:")
print(f"- Mínima: {expected.min():.2f}")
print(f"- Todas >= 5: {'Sí' if expected.min() >= 5 else 'No'}")

if expected.min() >= 5:
    print("- Los supuestos de la prueba Chi-cuadrado se cumplen")
else:
    print("-  Advertencia: Algunas frecuencias esperadas < 5")
```

```python
# Comparación de Métodos Estadísticos
print(f"\n" + "="*60)
print("COMPARACIÓN DE MÉTODOS ESTADÍSTICOS")
print("="*60)

print(f"\n1. PRUEBA Z DE PROPORCIONES:")
print(f"   - Estadístico z: {z_stat:.4f}")
print(f"   - Valor p: {p_value_conversion:.6f}")
print(f"   - Conclusión: {'Significativo' if p_value_conversion < alpha else 'No significativo'}")

print(f"\n2. PRUEBA CHI-CUADRADO:")
print(f"   - Estadístico χ²: {chi2_stat:.4f}")
print(f"   - Valor p: {p_value_chi2:.6f}")
print(f"   - Conclusión: {'Significativo' if p_value_chi2 < alpha else 'No significativo'}")

print(f"\n3. CONSISTENCIA ENTRE MÉTODOS:")
consistente = (p_value_conversion < alpha) == (p_value_chi2 < alpha)
print(f"   - Los métodos son consistentes: {'Sí' if consistente else 'No'}")

# 4. Relación matemática entre los métodos
print(f"\n4. RELACIÓN MATEMÁTICA:")
print(f"   - z² = χ²: {z_stat**2:.4f} = {chi2_stat:.4f}")
print(f"   - Diferencia: {abs(z_stat**2 - chi2_stat):.6f}")
print(f"   - Los estadísticos están relacionados por: χ² = z²")

print(f"\n" + "="*60)
print("INTERPRETACIÓN FINAL")
print("="*60)

if p_value_conversion < alpha:
    print("- RESULTADO: Diferencia estadísticamente significativa")
    mejor_variante = "Tratamiento" if tasa_tratamiento > tasa_control else "Control"
    print(f"   - La variante {mejor_variante} tiene mejor conversión")
    print(f"   - Diferencia: {abs(tasa_tratamiento - tasa_control):.2%}")
    print(f"   - Recomendación: Implementar la variante {mejor_variante}")
else:
    print("- RESULTADO: No hay diferencia estadísticamente significativa")
    print("   - Las tasas de conversión son similares")
    print("   - No se puede concluir que una variante sea mejor")
    print("   - Recomendación: Mantener versión actual o extender experimento")

print(f"\n RESUMEN TÉCNICO:")
print(f"   - Tamaño de muestra: {len(df):,} usuarios")
print(f"   - Control: {totales['control']:,} usuarios ({tasa_control:.2%} conversión)")
print(f"   - Tratamiento: {totales['tratamiento']:,} usuarios ({tasa_tratamiento:.2%} conversión)")
print(f"   - Nivel de confianza: {(1-alpha)*100}%")
print(f"   - Poder estadístico: Adecuado para detectar diferencias ≥ 1%")
```
